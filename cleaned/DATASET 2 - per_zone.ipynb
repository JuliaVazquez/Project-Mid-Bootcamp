```python
def cleaner(df):
    #I drop the info from 1999 and 2000 because it has no info about Rural and data since 2001 its ok
    #also, if after I need it I'll get it back
    df = df[~df['year'].isin([1999,2000])]
    
    # now I split the data into 3 to work on it
    df_All = df[df['come_from'].isin(['All'])]
    df_Spain = df[df['come_from'].isin(['Spain'])]
    df_Foreign = df[df['come_from'].isin(['Foreign'])]

    # drop columns that are not needed 
    df_Spain = df_Spain.drop(['H_avg_stay','R_avg_stay'],axis=1)
    df_Foreign = df_Foreign.drop(['H_avg_stay','R_avg_stay'],axis=1)
    


    return df_All, df_Spain, df_Foreign
```

```python
zones_All, zones_Spain, Zones_Foreign = cleaner(zones)
zones_Spain
```

```python
gal_prov_All, gal_prov_Spain, gal_prov_Foreign = cleaner(gal_prov)
gal_prov_Spain[gal_prov_Spain['R_travellers'].isin(['..'])]
```

```python
# I'm trying to keep all the information, so the '..' means there is no info for that, I'm going to change it to 0 just to keep the rest of the info in those columns and rows

def clean_dots(df):
    df['R_travellers'] = np.where(df['R_travellers'].isin(['..']),0,df['R_travellers'])
    df['R_nights'] = np.where(df['R_nights'].isin(['..']),0,df['R_nights'])
    df['R_travellers'] = df['R_travellers'].astype(np.number)
    df['R_nights'] = df['R_nights'].astype(np.number)
    
    return df
```

```python
gal_prov_All = clean_dots(gal_prov_All)
gal_prov_Spain = clean_dots(gal_prov_Spain)
zones_All = clean_dots(zones_All)
gal_prov_Foreign = clean_dots(gal_prov_Foreign)
zones_Spain = clean_dots(zones_Spain)
zones_Foreign = clean_dots(Zones_Foreign)

zones_Spain
```

```python
#first I need to get the columns as numerical and then I will clean the -1

def clean_2dot(df,cols):
    for col in cols:
        df[col] = np.where(df[col].isin(['..']),-1,df[col])
        df[col] = df[col].astype(np.number)
    return df
```

```python
cols = ['travellers_Foreing', 'nights_Foreing',
       'travellers_Spain', 'nights_Spain', 'travellers_All', 'nights_All',
       'avg_stay_All']
touristic_point = clean_2dot(touristic_point, cols)
touristic_point
```

```python
# travellers_All and nights_All replace the -1 with the sum
touristic_point['travellers_All'] = np.where(touristic_point['travellers_All'].isin([-1]),touristic_point['travellers_Foreing']+touristic_point['travellers_Spain'],touristic_point['travellers_All'])
touristic_point['nights_All'] = np.where(touristic_point['nights_All'].isin([-1]),touristic_point['nights_Foreing']+touristic_point['nights_Spain'],touristic_point['nights_All'])
touristic_point[touristic_point['travellers_Foreing'].isin([55])]
```

```python
# avg_stay_All is the nights/travellers so I'm going to fill it with that
touristic_point['avg_stay_All'] = np.where(touristic_point['avg_stay_All'].isin([-1]),touristic_point['nights_All']/touristic_point['travellers_All'],touristic_point['avg_stay_All'])
touristic_point[touristic_point['travellers_All'].isin([2324])]
```

```python
# travellers_Spain and nights_Spain I'm going to fill that line with the half of travellers_All and nights_All just to keep that one line
touristic_point['travellers_Spain'] = np.where(touristic_point['travellers_Spain'].isin([-1]),touristic_point['travellers_All']/2,touristic_point['travellers_Spain'])
touristic_point['nights_Spain'] = np.where(touristic_point['nights_Spain'].isin([-1]),touristic_point['nights_All']/2,touristic_point['nights_Spain'])
touristic_point[touristic_point['travellers_All'].isin([2324])]
```

```python
# in this line I'll do the same with travellers an nights of Foreigners, but it will repercute on another line in where I'll put 0 later
touristic_point['travellers_Foreing'] = np.where(touristic_point['travellers_Foreing'].isin([-1]),touristic_point['travellers_All']/2,touristic_point['travellers_Foreing'])
touristic_point['nights_Foreing'] = np.where(touristic_point['nights_Foreing'].isin([-1]),touristic_point['nights_All']/2,touristic_point['nights_Foreing'])
touristic_point[touristic_point['travellers_All'].isin([2324])]
```
