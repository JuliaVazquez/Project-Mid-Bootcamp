```python
data = pd.read_excel('Viajes, pernoctaciones, duración media y gasto por destino principal (extranjero y comunidades autónomas).xlsx')
data
```

```python
data = data.drop(['Destino'], axis=1)
```

```python
data['Concepto tur�stico'].value_counts()
```

```python
def by_year(df, year):
    df = df[df['Periodo'].isin([year])]
    return df
```

```python
data_2015 = by_year(data,2015)
data_2016 = by_year(data,2016)
data_2017 = by_year(data,2017)
data_2018 = by_year(data,2018)
data_2019 = by_year(data,2019)
data_2020 = by_year(data,2020)
data_2021 = by_year(data,2021)
```

```python
def clean_tables(df):
    df['valor_absoluto'] = np.where(df['Tipo de dato'].isin(['Valor absoluto']), df['Total'],0)
    df['distribucion_porcentual'] = np.where(df['Tipo de dato'].isin(['Distribuci�n porcentual']),df['Total'], 0 )
    df['variacion_anual'] = np.where(df['Tipo de dato'].isin(['Variaci�n anual']),df['Total'], 0 )
    df = df.drop(['Tipo de dato','Total','Periodo'],axis=1)
    df = df.groupby(['Concepto tur�stico']).agg({'valor_absoluto':max,'distribucion_porcentual':max,'variacion_anual':max})
    df = df.drop(['variacion_anual','distribucion_porcentual'],axis=1)
    return df
```

```python
data_2015 = clean_tables(data_2015)
data_2016 = clean_tables(data_2016)
data_2017 = clean_tables(data_2017)
data_2018 = clean_tables(data_2018)
data_2019 = clean_tables(data_2019)
data_2020 = clean_tables(data_2020)
data_2021 = clean_tables(data_2021)
```

```python
data = pd.concat([data_2015,data_2016,data_2017,data_2018,data_2019,data_2020,data_2021], axis=1)
data.columns = [2015,2016,2017,2018,2019,2020,2021]
data = data.T
data
```

```python
data.columns = ['avg_length_trip','avg_expenses_person_day','avg_expenses_person','total_expenses','total_nights','total_trips']
data
```

```python
data.to_csv('data_2015-2021_totals_avgs.csv', index=True)
```
