# Reading df0

```python
df0 = pd.read_excel('Viaxeiros, noites e estadía media en establecementos hoteleiros e de turismo rural. España, Galicia e provincias. Datos mensuaisexcel.xls')
```

```python
df0
```

## cleaning df0

```python
def clean_data(df):
    # droping NaN because they are residual rows
    df = df.dropna()
    
    # changing column names
    df.columns = ['destination','come_from','period','H_travellers','H_nights', 'H_avg_stay','R_travellers','R_nights','R_avg_stay']

    # changing España to Spain
    df['destination'] = np.where(df['destination'].isin(['España']), 'Spain', df['destination'] )

    #renaming 'come_from' values
    df['come_from'] = df['come_from'].map({'Residentes en España':'Spain', 'Residentes no estranxeiro':'Foreign', 'Todas as procedencias':'All'})
    
    #spliting period into 2 columns
    years = []
    months = []
    for i in df['period']:
        year = i.split('/')[0]
        years.append(year)
        month = i.split('/')[1]
        months.append(month)
    df['year'] = years
    df['month'] = months
    df = df.drop(['period'],axis=1)

    #cleaning names in month column
    df['month'] = df['month'].map({'Xaneiro':'Jan', 'Febreiro':'Feb','Marzo':'Mar','Abril':'Apr','Maio':'May','Xuño':'Jun', 'Xullo':'Jul','Agosto':'Aug','Setembro':'Sep','Outubro':'Oct','Novembro':'Nov','Decembro':'Dec'})

    return df
```

```python
df0 = clean_data(df0)
```

```python
df0.to_csv('sp_gal_prov_trav_nights_avg_stay.csv', index=True)
```
