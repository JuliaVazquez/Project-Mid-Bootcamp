```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
person_day = pd.read_csv('avg_per_person_and_day.csv')
dist_ca = pd.read_csv('dist_per_ca.csv')
dist_country = pd.read_csv('dist_per_country.csv')
```

```python
def drop0(df):
    df = df.drop(['Unnamed: 0'],axis=1)
    return df
```

```python
person_day = drop0(person_day)
dist_ca = drop0(dist_ca)
dist_country = drop0(dist_country)
```

```python
dist_country.columns = ['come_from','percentage_Galicia','year','trimester']
```

```python
dist_ca.columns = ['come_from','percentage_Spain','year','trimester']
```

```python
dist_ca
```

```python
person_day = person_day[['come_from','year','trimester','avg_person_day_€','avg_person_€']]
```

```python
def merge(df1, df2, how_to):
    merge = pd.merge(left = df1,
        right = df2,
        how = how_to, 
        left_on = 'come_from',
        right_on= 'come_from')
    merge = merge[(merge['year_x']==merge['year_y'])&(merge['trimester_x']==merge['trimester_y'])]
    merge = merge.drop(['year_y','trimester_y'],axis=1)
    merge
    return merge
```

```python
merge1 = merge(dist_country,person_day,'inner')
merge1.columns = ['come_from','percentage_Galicia','year','trimester','avg_person_day_€','avg_person_€']
merge0 = merge1[merge1['come_from'].isin(['All'])]
merge1 = merge1[~merge1['come_from'].isin(['All'])]
```

```python
merge2 = merge(merge1,dist_ca,'left')
merge2['come_from'].value_counts()
```

```python
 #cleaning names in come_from column
merge2['come_from'] = merge2['come_from'].map({'Alemania':'Germany','Argentina':'Argentina','Brasil':'Brasil','Chile':'Chile','Colombia':'Colombia','Andorra':'Andorra','Australia':'Australia','China':'China','Corea':'Korea','Rusia':'Russia','Portugal':'Portugal','Polonia':'Polonia','Austria':'Austria','M�xico':'Mexico','Canad�':'Canada','Estados Unidos de Am�rica':'USA','Resto de Am�rica':'Others America','Emiratos �rabes Unidos':'Emirates', 'Jap�n':'Japan','Turqu�a':'Turkey','Otros Pa�ses Europeos':'Others Europe','Suiza':'Switzerland','Suecia':'Sweden','Ruman�a':'Romania', 'Reino Unido':'UK','Pa�ses Bajos':'Netherlands','Noruega':'Norway','Italia':'Italy','Irlanda':'Ireland','Francia':'France','Finlandia':'Finland','Dinamarca':'Denmark','B�lgica':'Belgium','Resto del mundo':'Other'})
merge2['come_from'].value_counts(dropna=False)
```

```python
merge3 = pd.concat([merge0,merge2],axis=0)
merge3
```

```python
# the . in €_Galicia , €_Spain, avg_person_day_€ and avg_person_€ are from the 2º trimester of 2020 so I'm going to fill it with 0
merge2['percentage_Galicia'] = np.where(merge2['percentage_Galicia'].isin(['.']),0,merge2['percentage_Galicia'])
merge2['percentage_Spain'] = np.where(merge2['percentage_Spain'].isin(['.']),0,merge2['percentage_Spain'])
merge2['avg_person_day_€'] = np.where(merge2['avg_person_day_€'].isin(['.']),0,merge2['avg_person_day_€'])
merge2['avg_person_€'] = np.where(merge2['avg_person_€'].isin(['.']),0,merge2['avg_person_€'])
```

```python
# now converting these columns into numbers
merge2['percentage_Galicia'] = merge2['percentage_Galicia'].astype(np.number)
merge2['percentage_Spain'] = merge2['percentage_Spain'].astype(np.number)
merge2['avg_person_day_€'] = merge2['avg_person_day_€'].astype(np.number)
merge2['avg_person_€'] = merge2['avg_person_€'].astype(np.number)
```

```python
merge2.columns = ['come_from','percentage_Galicia','year','trimester','avg_person_day','avg_person','percentage_Spain']
merge2 = merge2[['come_from','year','trimester','avg_person_day','avg_person','percentage_Galicia','percentage_Spain']]
```

```python
merge2.to_csv('spent_per_country.csv', index=False)
```
