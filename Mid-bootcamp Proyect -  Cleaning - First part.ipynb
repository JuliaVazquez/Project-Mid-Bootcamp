```python
def clean_M(df,col):
    years = []
    M = []
    for i in df[col]:
        year = i.split('M')[0]
        years.append(year)
        month = i.split('M')[1]
        M.append(month)
    df['year'] = years
    df['month'] = M
    df = df.drop(['period'],axis=1)
    return df
```

```python
def clean_month(df):
    df['month'] = df['month'].map({'01':'Jan', '02':'Feb','03':'Mar','04':'Apr','05':'May','06':'Jun', '07':'Jul', '08':'Aug', '09':'Sep','10':'Oct','11':'Nov','12':'Dec'})
    return df
```

```python
t_totals = clean_month(clean_M(t_totals,'period'))
n_totals = clean_month(clean_M(n_totals,'period'))

t_per_province = clean_month(clean_M(t_per_province,'period'))
n_per_province = clean_month(clean_M(n_per_province,'period'))

t_per_accommodation = clean_month(clean_M(t_per_accommodation,'period'))
n_per_accommodation = clean_month(clean_M(n_per_accommodation,'period'))
```

```python
dist_per_country = pd.read_excel('Distribución del gasto en destino realizado según país de residencia2.xlsx')
dist_per_ca = pd.read_excel('Distribución del gasto en destino de cada país de residencia según comunidad autónoma2.xlsx')
```

```python


def rename_dist(df):
    df = df.drop(['Comunidades y Ciudades Aut�nomas de destino'],axis=1)
    df.columns = ['come_from', 'period', '%']

    return df
```

```python
# renaming columns

dist_per_country = rename_dist(dist_per_country)
dist_per_ca = rename_dist(dist_per_ca)
```

```python
# distribution inside Galicia

# dist_per_country = 
dist_per_country[~dist_per_country['come_from'].isin(['Total'])]
```

```python
# distribution pero ca, in relation with thew other ca's from spain

dist_per_ca
```

```python
def clean_T(df,col):
    years = []
    T = []
    for i in df[col]:
        year = i.split('T')[0]
        years.append(year)
        trim = i.split('T')[1]
        T.append(trim)
    df['year'] = years
    df['trimester'] = T
    df = df.drop(['period'],axis=1)
  
    return df
```

```python
# cleaning period column where has Trimesters and Years

dist_per_ca = clean_T(dist_per_ca,'period')
dist_per_country = clean_T(dist_per_country,'period')
```

```python
# renaming just a few All

dist_per_country['come_from'] = np.where(dist_per_country['come_from'].isin(['Total']), 'All', dist_per_country['come_from'] )
```

```python
dist_per_ca.to_csv('dist_per_ca.csv', index=True)
dist_per_country.to_csv('dist_per_country.csv', index=True)
```
